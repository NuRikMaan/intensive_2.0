```python
'''удалении аномалии Мы реализовали два подхода
Z-Score: Использует среднее и стандартное отклонение. Значения, которые находятся на заданное количество стандартных отклонений от среднего, считаются аномальными.
IQR: Использует 25-й и 75-й процентили. Значения, которые выходят за рамки заданного диапазона, вычисляемого с помощью IQR, считаются аномальными.
файл valid.csv обратотан с помощью метода Z-score'''
```

```python
import pandas as pd
import numpy as np

def remove_outliers(df, method='iqr', threshold=1.5):
    
    df_filtered = df.copy()
    removed_counts = {}
    for column in df.select_dtypes(include=np.number).columns:
        if method == 'zscore':
            z = np.abs((df[column] - df[column].mean()) / df[column].std())
            outliers = (z > threshold)
        elif method == 'iqr':
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR
            outliers = ((df[column] < lower_bound) | (df[column] > upper_bound))
        else:
            raise ValueError("Неверный метод. Допустимые методы: 'zscore', 'iqr'.")
        
        removed_count = outliers.sum()
        removed_counts[column] = removed_count
        df_filtered = df_filtered[~outliers]

    return df_filtered, removed_counts

# 1. Загрузка данных
try:
    input_file = 'valid.csv'
    df = pd.read_csv(input_file)
    print("Файл успешно загружен.")
except FileNotFoundError:
    print(f"Ошибка: Файл {input_file} не найден.")
    exit()
except Exception as e:
    print(f"Произошла ошибка при чтении файла: {e}")
    exit()

# 2. Подсчет количества строк перед удалением
initial_rows = len(df)


# 3. Удаление аномалий (z-score)
method = 'zscore'
threshold = 3
df_no_outliers, removed_counts = remove_outliers(df, method, threshold)
final_rows = len(df_no_outliers)

# 4. Вывод информации
print(f"\nКоличество строк до удаления аномалий: {initial_rows}")
print(f"Количество строк после удаления аномалий: {final_rows}")
print("Количество удаленных аномалий по столбцам:")
for column, count in removed_counts.items():
    print(f"  {column}: {count}")

# 5. Сохранение результата
try:
    df_no_outliers.to_csv('valid_3.csv', index=False)
    print("\nДанные без выбросов сохранены в valid_1.csv.")
except Exception as e:
    print(f"Ошибка при сохранении файла: {e}")
    exit()

print("\nГотово.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_missing_data(filepath, image_filename):
   

    # Загружаем данные из CSV
    try:
        df = pd.read_csv(filepath)
    except FileNotFoundError:
        print(f"Ошибка: Файл не найден: {filepath}")
        return
    except Exception as e:
        print(f"Ошибка при чтении файла: {e}")
        return

    # Выявляем пропуски в данных
    missing_data = df.isnull()

    # Создаем тепловую карту (heatmap) для визуализации пропусков
    plt.figure(figsize=(12, 8))  # Задаём размер графика
    sns.heatmap(missing_data, cbar=False, cmap="viridis") # убрали color bar и изменили цвет
    plt.title(f"Распределение пропущенных значений в {filepath}", fontsize=16) # задали имя заголовка и размер
    plt.xlabel("Колонки", fontsize=14) # задали имя оси и размер
    plt.ylabel("Индексы строк", fontsize=14) # задали имя оси и размер
    plt.savefig(image_filename, bbox_inches='tight') # задали имя сохранения и подогнади размеры

    # Отображаем график
    plt.show()


# Применяем функцию к обоим файлам
file1_path = "valid_1.csv"  # Замените на свой путь
file2_path = "train_1.csv"  # Замените на свой путь
image1_filename = "missing_data_file1.png"
image2_filename = "missing_data_file2.png"


visualize_missing_data(file1_path, image1_filename)
visualize_missing_data(file2_path, image2_filename)
```

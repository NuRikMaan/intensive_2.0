```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('train.csv')  #

print(df.head())  # Показывает первые 5 строк
print(df.info())  # Общая информация о DataFrame
print(df.describe()) # Статистика по числовым признакам
print(df.shape) # (количество строк, количество столбцов)
print(df.columns)
print(df.dtypes)
print(df.isnull().sum()) # Показывает количество NaN в каждом столбце
for col in df.select_dtypes(include=['object']):
    print(f"\n{col}: {df[col].nunique()} unique values")
    print(df[col].unique())
    
df.hist(figsize=(15, 15))
plt.show()
```
